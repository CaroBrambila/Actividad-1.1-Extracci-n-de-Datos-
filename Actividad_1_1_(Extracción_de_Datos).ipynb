```python
#Filtro y
filtro7 = df[(df['SUBTOTAL_PARTIDA'] < 1000) & (df['FECHA_DOC'] == "19/01/2022  12:00:00 a. m.")]
filtro7

```

```python
# filtro o
filtro8=df[(df["MARGEN_UNITARIO_CALCULADO"] < 6000)| (df["FECHA_DOC"]=="05/01/2022  12:00:00 a. m.")]
filtro8
```

```python
#filtro not
filtro9=df[~(df["SUBTOTAL_PARTIDA"] > 1000) & ~(df["FECHA_DOC"]=="08/09/2022  12:00:00 a. m.")]
filtro9
```
